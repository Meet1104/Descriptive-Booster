1. Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/apple/Desktop/MATH/household_dataset_200.csv")
df.head()
```

2. Identify Data Types

```python
df.dtypes
```

3. Central Tendency

```python
print("Central Tendency of income:-")
print("Mean:",df['Household_Income'].mean())
print("Median:",df['Household_Income'].median())
print("Mode:",df['Household_Income'].mode()[0])

print("\nCentral Tendency of income:-")
print("Mean:",df['Age_of_Household_Head'].mean())
print("Median:",df['Age_of_Household_Head'].median())
print("Mode:",df['Age_of_Household_Head'].mode()[0])
```

4.Measures of Dispersion

```python
print("Measures of Dispersion for income:-")
print("Range", df['Household_Income'].max() - df['Household_Income'].min())
print("Variance:", df['Household_Income'].var())
print("Standard Deviation:", df['Household_Income'].std())

print("\nInterquartile Range (IQR) for income:-")
print("IQR:", df['Household_Income'].quantile(0.75) - df['Household_Income'].quantile(0.25))
```

5.Distribution & Visualization

```python
sns.histplot(df['Household_Income'], kde=True)
plt.title('Distribution of Household Income')
plt.show()
```

6.Skewness & Kurtosis

```python
df['Household_Income'].skew()
df['Household_Income'].kurtosis()
```

7.Data Categorization (Income vs Education / Area)

```python
print("Boxplot for Household Income by Education Level:")
sns.boxplot(x='Education_Level', y='Household_Income', data=df)
plt.title('Household Income by Education Level')
plt.show()

print("\nBoxplot for Household Income by Area Type:")
sns.boxplot(x='Urban_Rural', y='Household_Income', data=df)
plt.title('Household Income by Area Type')
plt.show()
```
